Set Operations

```python
ts = {20, 40, "World", 40}
```

```python
ts1 ={1,2,3,40}
```

```python
print(ts)
ts1
```

```python
ts.add(50)
ts
```

```python
ts.copy()
```

```python
ts.difference(ts1)
# returns the value present in 1st set but not in 2nd set
```

```python
ts.difference_update(ts1)
ts
```

```python
ts.discard("World")
ts
```

```python
ts
```

```python
ts.intersection(ts1)
```

```python
# Intersection: Elements present in both sets
```

```python
ts.intersection_update(ts1)
ts1
```

```python
# Intersection_update: Keeps only comman elements in set and remove all other elements from the original set
```

```python
ts.isdisjoint(ts1)
```

```python
# isdisjoint: returns true if elements have common elements and false if elements do not have any common element
```

```python
ts.issubset(ts1)
```

```python
# issubset: need to have all elements of 1st set into 2nd
```

```python
ts.issuperset(ts1)
```

```python
# issuperset: needs to have all elements of 2nd set into 1st
```

```python
ts.pop()
```

```python
ts.symmetric_difference(ts1)
```

```python
# removes common elements from sets and returns remaining elements
```

```python
ts.symmetric_difference_update(ts1)
ts
```

```python
# returns remaining values in 1st set
```

```python
ts.union(ts1)
```

```python
ts.update()
ts
```
