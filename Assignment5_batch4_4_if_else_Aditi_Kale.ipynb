### 42. Check if a number is perfect (sum of factors equals number).

```python
num = int(input("Enter a number:"))
sum_factor=0
for i in range(1,num):
    if num%i==0:
        sum_factor+=i
if sum_factor == num:
    print(num,"is a perfect number")
else:
    print(num,"is not a perfect number")
```
