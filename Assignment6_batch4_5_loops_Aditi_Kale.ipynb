### 40. Create a dictionary of 5 products and their prices. Increase every product's price by 10% using a for loop

```python
products = {"chair": 2000,"table": 3000,"fridge": 10000,"TV": 20000,"sofa": 15000}

for key in products:
    inc = products[key] * (10 / 100)  
    products[key] = int(products[key] + inc) 

print(products)
```

### 41. Given a dictionary of 10 items, delete all items whose value is less than 50 using a for loop

```python
products = {"chair": 200,"table": 30,"fridge": 100,"TV": 20,"sofa": 150,"bed": 18,"fan": 25,"lamp": 12,"mixer": 35,"oven": 80}
for key,value in list(products.items()):
    if value<50:
        products.pop(key)
print(products)
```
